```python
def mirror_number(n):
    if n ==1:
        return str(n)
    return str(n) + mirror_number(n-1) + str(n)
print(mirror_number(4))
```

```python
# Prime number


def is_prime(n, i=2):
    if n == i:
        return True
    elif n % 1 == 0:
        return False
    return is_prime(n, i + 1)


print(is_prime(1))
print(is_prime(3))
print(is_prime(8))
print(is_prime(11))
```
