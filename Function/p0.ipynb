# Function:

- A function is a reusable block of code that performs a specific task.

- Function is container which contain the business or application logic.

- it is a name given to the memory addition which contain the business logic.

- Instead of writing the same code again and again, you write it once inside a function and call it whenever needed.

# 🔹 Why use Functions?

- Code reusability

- Easy to read and manage

- Reduces repetition

- Helps in modular programming

- Easier debugging

 # utility function: 

 - A utility function is a small helper function created to perform a common or repeated task in a program.

- It is not the main logic of the program, but it supports other functions by reducing repeated code.

# function of string

- String functions are built-in methods used to perform operations on strings like changing case, searching, replacing, splitting, checking, etc.

- In Python, strings are immutable (cannot be changed directly), so string functions return a new string.


#  ⭐ Most Important String Functions :

| Function       | Use                                        |
| -------------- | ------------------------------------------ |
| `upper()`      | Convert string to uppercase                |
| `lower()`      | Convert string to lowercase                |
| `split()`      | Convert string → list                      |
| `join()`       | Convert list → string                      |
| `replace()`    | Replace text in string                     |
| `find()`       | Search text and return index               |
| `strip()`      | Remove spaces from start and end           |
| `count()`      | Count occurrences of characters/substrings |
| `startswith()` | Check if string starts with given text     |
| `endswith()`   | Check if string ends with given text       |
