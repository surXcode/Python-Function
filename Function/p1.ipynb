```python
s = input("Enter a string:")

upp = ""
low = ""
digits = ""
symbols = ""

for ch in s:
    if ch.isupper():
        upp += ch
    elif ch.islower():
        low += ch
    elif ch.isdigit():
        digits += ch
    else:
        symbols += ch

print(upp)
print(low)
print(digits)
print(symbols)
```

```python
# Extract all the palindrome words from a string.

s = "This is beginer level eye catching phone"
w = s.split()
palin = [w for w in w if w.lower() == w.lower()[::-1]]
print(palin)
```

```python
# find longest substring
 
s = "This is beginer level eye catching phone"
longest_word = max(s.split(), key=len)
print(longest_word)
```

```python
# Find nth prime number

def nth_prime(n):
    count = 0
    num = 2
    
    while True:
        prime = True
        
        for i in range(2, num):
            if num % i == 0:
                prime = False
                break
        
        if prime:
            count += 1
        
        if count == n:
            return num
        
        num += 1
n = int(input("Enter value of n: "))
print(f"{n}th prime number is:", nth_prime(n))
```

```python
def nth_prime(n):
    count = 0
    num = 1
    
    while count < n:
        num += 1
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            count += 1
            
    return num


n = int(input("Enter value of n: "))
print("Nth prime number is:", nth_prime(n))
```
