# Argument:

- Argument is the value that you pass to a function when calling it.

👉 Parameter = variable in function definition.

👉 Argument = actual value passed to the function.

# Types of argument:

- There are 5 main types of arguments:

1. Positional Arguments

2. Keyword Arguments

3. Default Arguments

4. Variable Length Arguments (*args)

5. Keyword Variable Length Arguments (**kwargs)

# 1. Positional Arguments:

- Th arguments which are declare or created in function declaration are called as positional argument.

```python
def sam(a, b):
    print(a)
    print(b)
sam(55,44)
```

# 3. Default Arguments:

- Parameter has a default value.

🧠 How it works:

- If value is not provided → default value is used

- If value is provided → default is overridden

```python
def sam(a, b=0):
    print(a)
    print(b)
sam(55)
```

# 2. Keyword Arguments-

- Passing key value pair as the value to the arguments. 
- they are present in function call

🧠 How it works:

- Values assigned using parameter name

- Order does NOT matter

```python
def sam(**d):
    print(d)
sam(a=2,b=9,c=55)
```

# 4. Variable Length Arguments (*args):

- These arguments are capable to taking values from 1 to n number of inputs or values.

- They are present in function declaration.

- accept n number of positional argument.

- store the value in tuple.

🧠 How it works:

- Collects extra positional arguments

- Stored as a tuple

```python
# args & kwargs


def sam(*args, **kwargs):
    print(args)
    print(kwargs)
    print(type(args))
    print(type(kwargs))
sam(2,3,4,5,67,11,a=2,b=True,c='hello',d=3.4)
```

| Type        | Stored As        | Order Important | Flexible |
|------------|------------------|----------------|----------|
| Positional | Normal Variable  | ✅ Yes         | ❌ No    |
| Keyword    | Normal Variable  | ❌ No          | ❌ No    |
| Default    | Normal Variable  | ❌ No          | ❌ No    |
| *args      | Tuple            | ✅ Yes         | ✅ Yes   |
| **kwargs   | Dictionary       | ❌ No          | ✅ Yes   |

```python
# create a function which can extract all the palindrome values from a list


def filter_palin(list1):

    def is_palin(s):
        return s == s[::-1]

    return [s for s in list1 if is_palin(s)]
filter_palin([121, 123, "level", "python", 454, "madam", 789])
```

```python
# create a function which gives the factorial of number.

def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact
num = int(input("Enter a number: "))
print("Factorial is:", factorial(num))
```
