# Packing:

- It is a phenomenon of grouping the individual in the form of collection to provide security.

# Types:

1. Tuple apacking/ Single Packing
2. Dictionary Packing / Double Packing


# 1. Tuple apacking/ Single Packing:

- It is a phenomenon of grouping the individual values in the form of tuple collection to provide security.

- Syntax:

    - def fname(*args):
      - SB
    - fname(val1,val2,........,valn)

```python
def sam(*a):
    print('hello')
    print(a)
    print(type(a))
sam(22,33,44,54,66)
```

```python
a = 10, 20, 30
print(a)
print(type(a))
```

```python
data = "Suraj", 25, 95.5, True
print(data)
```

# 2. Dictionary Packing / Double Packing:

- It is phenomenon of grouping the individual key value pair in the form of dictionary.

* Note-

- key should follow the rules of identifiers.
- key should be of string data type only.
- key should'nt enclose with '' but values must be enclosed with '' if it is string.

# Syntax:

- def fname(** var):
    - SB
- fname(k1=v1,k2=v2,......,kn=vn)

```python
def sam(**a):
    
    print(a)
    print(type(a))
sam(a=2,b=True,c='hii',d=3.4)
```

```python
# args & kwargs


def sam(*args, **kwargs):
    print(args)
    print(kwargs)
    print(type(args))
    print(type(kwargs))
sam(2,3,4,5,67,11,a=2,b=True,c='hello',d=3.4)
```

# Unpacking:

- It is phenomenon of dividing the collection and storing each value in unique variable.

```python
# a, b, c, d = 'mnop'
a, b, c, d = [1,4,5,6]
print(a)
print(b)
print(c)
print(d)
```

# Summary 

| Concept        | Meaning                                      |
|---------------|----------------------------------------------|
| Packing       | Multiple values → One variable               |
| Unpacking     | One collection → Multiple variables          |
| *args         | Packs into tuple                            |
| **kwargs      | Packs into dictionary                       |
| *variable     | Extended unpacking                          |

## 🔥 Quick Summary Table

| Type                  | Symbol Used | Data Type    |
|-----------------------|------------|-------------|
| Tuple Packing         | ,          | Tuple       |
| Tuple Unpacking       | =          | Tuple       |
| Extended Unpacking    | *          | List        |
| Function Packing      | *args      | Tuple       |
| Dictionary Packing    | **kwargs   | Dictionary  |
| Dictionary Unpacking  | **         | Dictionary  |

```python
def sam(a, b, c, d):
    print(a, b, c, d)
sam(*'pqrs')
```

```python
def sam(a, b, c, d):
    print(a, b, c, d)

sam(**{'a': 33, 'b': 22, 'c': 55, 'd': 43})
```

```python
def sam(a, b, c, d):
    print(a, b, c, d)
sam(*{'m': 33, 'n': 22, 'o': 55, 'p': 43}.values())
```

```python
def sam(a, b, c, d):
    print(a, b, c, d)
sam(*{'m': 33, 'n': 22, 'o': 55, 'p': 43}.())
```
